# Pipeline Evaluation Dashboard

**Objective:** Compare three pipelines on a 200-call labeled dataset

| Pipeline | Metric Predicted | Architecture |
|----------|-----------------|-------------|
| **Unified** (Hybrid RAG) | PCS + NPS | BGE-M3 + RAPTOR + HyDE + GPT-5-mini rationale |
| **NPS_LLM** | NPS | Direct GPT-5-mini structured output |
| **PCS_predictive** (XGBoost) | PCS | GPT-5-mini feature extraction + XGBoost regression |

**Evaluations:**
1. Score distribution histograms (predicted vs ground truth)
2. Class distribution bar charts
3. Comparison table (runtime, cost per transcript, macro-averaged precision)
4. Token metrics table (avg/min/max per transcript)

```python
# --- Imports ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ast
import os
from sklearn.metrics import precision_score, classification_report

# Styling
sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams.update({'figure.dpi': 120, 'font.size': 10})

# Output directory for saved figures
FIG_DIR = './evaluation_figures'
os.makedirs(FIG_DIR, exist_ok=True)

print('Imports complete.')
```

```python
# --- Data Loading ---
DATA_DIR = './Hybrid RAG/files/eval_files'
RESULTS_DIR = './results_tables'

# Ground truth
gt_df = pd.read_csv(f'{DATA_DIR}/eval_table709.csv')

# Pipeline results
unified_df = pd.read_csv(f'{RESULTS_DIR}/hybrid_rag_results5.csv')
nps_llm_df = pd.read_csv(f'{RESULTS_DIR}/NPS_LLMDriver_Results.csv')
pcs_xgb_df = pd.read_csv(f'{RESULTS_DIR}/pcs_xgboost_output709.csv')

print(f'Ground truth:     {gt_df.shape}')
print(f'Unified:          {unified_df.shape}')
print(f'NPS_LLM:          {nps_llm_df.shape}')
print(f'PCS_predictive:   {pcs_xgb_df.shape}')
```

```python
# --- Validate alignment ---
gt_ids = set(gt_df['AgentRecordingSessionID'])
unified_ids = set(unified_df['AgentRecordingSessionID'])
nps_llm_ids = set(nps_llm_df['AgentRecordingSessionID'])
pcs_xgb_ids = set(pcs_xgb_df['AgentRecordingSessionID'])

common_ids = gt_ids & unified_ids & nps_llm_ids & pcs_xgb_ids
print(f'Common transcript IDs across all files: {len(common_ids)}')

if len(common_ids) < 200:
    print(f'  WARNING: Missing IDs detected')
    print(f'  GT only:      {len(gt_ids - common_ids)}')
    print(f'  Unified only: {len(unified_ids - common_ids)}')
    print(f'  NPS_LLM only: {len(nps_llm_ids - common_ids)}')
    print(f'  PCS_XGB only: {len(pcs_xgb_ids - common_ids)}')
```

## 1. Score Extraction & Class Mapping

```python
# --- Class mapping functions ---
def pcs_score_to_class(score):
    """Map PCS numeric score (0-10) to class."""
    if pd.isna(score):
        return None
    score = int(round(float(score)))
    if score <= 3:
        return 'Negative'
    elif score <= 6:
        return 'Neutral'
    else:
        return 'Positive'

def nps_score_to_class(score):
    """Map NPS numeric score (0-10) to class."""
    if pd.isna(score):
        return None
    score = int(round(float(score)))
    if score <= 4:
        return 'Detractor'
    elif score <= 6:
        return 'Passive'
    else:
        return 'Promoter'

print('Class mapping functions defined.')
```

```python
# --- Extract token data from PCS_XGBoost debug_info ---
def parse_debug_info(debug_str):
    """Parse debug_info string dict into usage values."""
    try:
        d = ast.literal_eval(debug_str)
        usage = d.get('usage', {})
        return pd.Series({
            'prompt_tokens': usage.get('prompt_tokens', 0),
            'completion_tokens': usage.get('completion_tokens', 0),
            'cached_tokens': usage.get('cached_tokens', 0),
            'total_tokens': usage.get('total_tokens', 0),
        })
    except:
        return pd.Series({
            'prompt_tokens': 0,
            'completion_tokens': 0,
            'cached_tokens': 0,
            'total_tokens': 0,
        })

# Parse PCS_XGBoost tokens
xgb_tokens = pcs_xgb_df['debug_info'].apply(parse_debug_info)
pcs_xgb_df = pd.concat([pcs_xgb_df, xgb_tokens], axis=1)

# NPS_LLM already has token columns from extract_token_data
# Ensure they're numeric
for col in ['prompt_tokens', 'completion_tokens', 'cached_tokens', 'total_tokens']:
    if col in nps_llm_df.columns:
        nps_llm_df[col] = pd.to_numeric(nps_llm_df[col], errors='coerce').fillna(0)

print(f'PCS_XGB token columns added: {list(xgb_tokens.columns)}')
print(f'NPS_LLM token columns: {[c for c in nps_llm_df.columns if "token" in c]}')
```

```python
# --- Build unified comparison DataFrame ---

# Ground truth scores and classes
eval_df = gt_df[['AgentRecordingSessionID', 'pcs_consensus_score', 'nps_consensus_score']].copy()
eval_df.columns = ['arsid', 'pcs_gt_score', 'nps_gt_score']
eval_df['pcs_gt_class'] = eval_df['pcs_gt_score'].apply(pcs_score_to_class)
eval_df['nps_gt_class'] = eval_df['nps_gt_score'].apply(nps_score_to_class)

# Unified pipeline predictions
unified_merge = unified_df[['AgentRecordingSessionID', 'pcs_score', 'pcs_class',
                            'nps_score', 'nps_class', 'token_count', 'processing_time_s']].copy()
unified_merge.columns = ['arsid', 'unified_pcs_score', 'unified_pcs_class',
                         'unified_nps_score', 'unified_nps_class',
                         'unified_token_count', 'unified_processing_time_s']
eval_df = eval_df.merge(unified_merge, on='arsid', how='inner')

# NPS_LLM predictions
nps_llm_merge = nps_llm_df[['AgentRecordingSessionID', 'score', 'class',
                            'prompt_tokens', 'completion_tokens', 'total_tokens']].copy()
nps_llm_merge.columns = ['arsid', 'nps_llm_score', 'nps_llm_class',
                         'nps_llm_prompt_tokens', 'nps_llm_completion_tokens',
                         'nps_llm_total_tokens']
eval_df = eval_df.merge(nps_llm_merge, on='arsid', how='inner')

# PCS_XGBoost predictions (features extracted, need XGBoost score)
# The XGBoost pipeline produces features but we need its predicted PCS score.
# Since the pipeline was already run, we'll use the feature data to apply the
# trained model. But first, let's check if there's already a prediction column.
print(f'PCS_XGB columns with "score" or "predict": '
      f'{[c for c in pcs_xgb_df.columns if "score" in c.lower() or "predict" in c.lower()]}')
```

```python
# --- PCS_XGBoost: Train model on extracted features and predict ---
# The pcs_xgboost_output709.csv has LLM-extracted features + consensus labels.
# We'll train XGBoost on the full dataset and use predictions (matching the pcs_predictive pipeline).

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline as SkPipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from xgboost import XGBRegressor

# Feature columns (matching pcs_predictive.ipynb)
feature_cols = [
    'pcs_features__agent_resolution_quality',
    'pcs_features__communication_quality',
    'pcs_features__ownership_level',
    'pcs_features__customer_effort_level',
    'pcs_features__agent_greeted_professionally',
    'pcs_features__agent_showed_empathy',
    'pcs_features__agent_acknowledged_customer_emotion',
    'pcs_features__agent_took_ownership',
    'pcs_features__agent_apologized_when_appropriate',
    'pcs_features__agent_explained_next_steps',
    'pcs_features__agent_set_expectations',
    'pcs_features__agent_used_clear_language',
    'pcs_features__agent_required_repeat_information',
    'pcs_features__agent_interrupted_customer',
    'pcs_features__agent_followed_process',
    'pcs_features__agent_deescalated_effectively',
    'pcs_features__agent_escalated_or_transferred_appropriately',
    'pcs_features__agent_closed_professionally',
    'pcs_features__agent_professionalism',
    'confidence__feature_extraction_confidence',
]

categorical_cols = [
    'pcs_features__agent_resolution_quality',
    'pcs_features__communication_quality',
    'pcs_features__ownership_level',
    'pcs_features__customer_effort_level',
]
numeric_cols = [c for c in feature_cols if c not in categorical_cols]

# Ordinal mappings (matching pcs_predictive.ipynb)
mappings = {
    'pcs_features__agent_resolution_quality': {
        'unknown / unclear': 0, 'unknown': 0, 'poor': 1, 'adequate': 2, 'good': 3, 'excellent': 4
    },
    'pcs_features__communication_quality': {
        'unknown / unclear': 0, 'unknown': 0, 'unclear': 1, 'acceptable': 2, 'clear': 3, 'exceptional': 4
    },
    'pcs_features__ownership_level': {
        'unknown / unclear': 0, 'unknown': 0, 'low': 1, 'moderate': 2, 'strong': 3
    },
    'pcs_features__customer_effort_level': {
        'unknown / unclear': 0, 'unknown': 0, 'high': 1, 'medium': 2, 'low': 3
    }
}

# Prepare features
xgb_model_df = pcs_xgb_df.copy()
for col, mapping in mappings.items():
    if col in xgb_model_df.columns:
        xgb_model_df[col] = xgb_model_df[col].astype(str).str.lower().str.strip().map(mapping).fillna(0).astype(int)

# Ensure numeric columns are numeric
for col in numeric_cols:
    if col in xgb_model_df.columns:
        xgb_model_df[col] = pd.to_numeric(xgb_model_df[col], errors='coerce').fillna(0)

# Target
target_col = 'pcs_consensus_score'
xgb_model_df[target_col] = pd.to_numeric(xgb_model_df[target_col], errors='coerce')
valid_mask = xgb_model_df[target_col].notna()

X = xgb_model_df.loc[valid_mask, feature_cols]
y = xgb_model_df.loc[valid_mask, target_col]

# Train XGBoost with best params from pcs_predictive.ipynb
xgb_model = XGBRegressor(
    objective='reg:squarederror',
    learning_rate=0.05,
    max_depth=5,
    n_estimators=100,
    subsample=0.8,
    random_state=42,
    n_jobs=-1
)
xgb_model.fit(X, y)

# Predict on all rows (full dataset prediction for evaluation)
xgb_preds = xgb_model.predict(xgb_model_df.loc[valid_mask, feature_cols])
xgb_model_df.loc[valid_mask, 'xgb_predicted_score'] = xgb_preds

print(f'XGBoost trained on {len(X)} samples')
print(f'Prediction range: [{xgb_preds.min():.2f}, {xgb_preds.max():.2f}]')
```

```python
# --- Merge PCS_XGBoost predictions into eval_df ---
xgb_merge = xgb_model_df[['AgentRecordingSessionID', 'xgb_predicted_score',
                           'prompt_tokens', 'completion_tokens', 'total_tokens']].copy()
xgb_merge.columns = ['arsid', 'pcs_xgb_score',
                     'pcs_xgb_prompt_tokens', 'pcs_xgb_completion_tokens',
                     'pcs_xgb_total_tokens']

# Round XGBoost predictions to nearest integer for class mapping
xgb_merge['pcs_xgb_score_rounded'] = xgb_merge['pcs_xgb_score'].round().astype(int).clip(0, 10)
xgb_merge['pcs_xgb_class'] = xgb_merge['pcs_xgb_score_rounded'].apply(pcs_score_to_class)

eval_df = eval_df.merge(xgb_merge, on='arsid', how='inner')

# Convert numeric prediction columns
for col in ['unified_pcs_score', 'unified_nps_score', 'nps_llm_score']:
    eval_df[col] = pd.to_numeric(eval_df[col], errors='coerce')

# Standardize Unified class names (ensure consistency)
eval_df['unified_pcs_class'] = eval_df['unified_pcs_score'].apply(pcs_score_to_class)
eval_df['unified_nps_class'] = eval_df['unified_nps_score'].apply(nps_score_to_class)
eval_df['nps_llm_class'] = eval_df['nps_llm_score'].apply(nps_score_to_class)

print(f'Final eval_df shape: {eval_df.shape}')
print(f'\nColumns: {list(eval_df.columns)}')
eval_df[['arsid', 'pcs_gt_score', 'pcs_gt_class', 'unified_pcs_score', 'unified_pcs_class',
         'pcs_xgb_score', 'pcs_xgb_class']].head()
```

## 4. Comparison Table: Runtime, Cost, Precision

**Cost model:** Azure OpenAI gpt-5-mini pricing (per 1M tokens).

**Note:** Unified pipeline token costs reflect full pipeline (PCS + NPS combined) since per-metric
API token breakdowns are not available in the results data. The Unified pipeline also uses BGE-M3
embedding compute which is not reflected in API token costs.

```python
# --- Pricing constants (Azure OpenAI gpt-5-mini) ---
# Update these if pricing changes
INPUT_PRICE_PER_1M = 0.40    # $ per 1M input tokens
OUTPUT_PRICE_PER_1M = 1.60   # $ per 1M output tokens
CACHED_INPUT_PRICE_PER_1M = 0.20  # $ per 1M cached input tokens

def compute_cost(prompt_tokens, completion_tokens, cached_tokens=0):
    """Compute cost per transcript in USD."""
    uncached_input = prompt_tokens - cached_tokens
    cost = (
        (uncached_input / 1_000_000) * INPUT_PRICE_PER_1M +
        (cached_tokens / 1_000_000) * CACHED_INPUT_PRICE_PER_1M +
        (completion_tokens / 1_000_000) * OUTPUT_PRICE_PER_1M
    )
    return cost

print(f'Pricing: ${INPUT_PRICE_PER_1M}/1M input, ${OUTPUT_PRICE_PER_1M}/1M output, ${CACHED_INPUT_PRICE_PER_1M}/1M cached')
```

```python
# --- Compute macro-averaged precision for each pipeline ---

def macro_precision(y_true, y_pred, labels):
    """Compute macro-averaged precision, handling missing classes."""
    # Filter to valid predictions only
    mask = pd.notna(y_true) & pd.notna(y_pred)
    y_true_clean = y_true[mask]
    y_pred_clean = y_pred[mask]
    return precision_score(y_true_clean, y_pred_clean, labels=labels,
                          average='macro', zero_division=0)

# PCS precision
pcs_precision_unified = macro_precision(
    eval_df['pcs_gt_class'], eval_df['unified_pcs_class'], pcs_classes)
pcs_precision_xgb = macro_precision(
    eval_df['pcs_gt_class'], eval_df['pcs_xgb_class'], pcs_classes)

# NPS precision
nps_precision_unified = macro_precision(
    eval_df['nps_gt_class'], eval_df['unified_nps_class'], nps_classes)
nps_precision_nps_llm = macro_precision(
    eval_df['nps_gt_class'], eval_df['nps_llm_class'], nps_classes)

print(f'PCS Macro Precision - Unified: {pcs_precision_unified:.3f}, XGBoost: {pcs_precision_xgb:.3f}')
print(f'NPS Macro Precision - Unified: {nps_precision_unified:.3f}, NPS_LLM: {nps_precision_nps_llm:.3f}')
```

```python
# --- Compute cost and runtime metrics ---

# NPS_LLM: per-transcript cost from token columns
nps_llm_cached = pd.to_numeric(nps_llm_df['cached_tokens'], errors='coerce').fillna(0)
nps_llm_cost_per_transcript = compute_cost(
    nps_llm_df['prompt_tokens'].mean(),
    nps_llm_df['completion_tokens'].mean(),
    nps_llm_cached.mean()
)

# PCS_XGBoost: per-transcript cost from parsed debug_info
pcs_xgb_cached = pd.to_numeric(pcs_xgb_df['cached_tokens'], errors='coerce').fillna(0)
pcs_xgb_cost_per_transcript = compute_cost(
    pcs_xgb_df['prompt_tokens'].mean(),
    pcs_xgb_df['completion_tokens'].mean(),
    pcs_xgb_cached.mean()
)

# Unified: no per-call token data; report processing time only
# Cost cannot be split; report as N/A or estimate from total time
unified_avg_time = eval_df['unified_processing_time_s'].mean()

# Runtime estimates
# NPS_LLM: extract processing_ms from debug_dict if available
if 'debug_dict' in nps_llm_df.columns:
    nps_llm_df['processing_ms_val'] = nps_llm_df['debug_dict'].apply(
        lambda x: ast.literal_eval(x).get('metadata', {}).get('processing_ms') if pd.notna(x) else None
    )
else:
    # Parse from debug_info string
    def get_processing_ms(debug_str):
        try:
            d = ast.literal_eval(debug_str)
            return d.get('metadata', {}).get('processing_ms')
        except:
            return None
    nps_llm_df['processing_ms_val'] = nps_llm_df['debug_info'].apply(get_processing_ms)

# PCS_XGB processing_ms
pcs_xgb_df['processing_ms_val'] = pcs_xgb_df['debug_info'].apply(
    lambda x: ast.literal_eval(x).get('metadata', {}).get('processing_ms') if pd.notna(x) else None
)

# Convert to seconds (processing_ms may be None for some Azure deployments)
nps_llm_runtime = pd.to_numeric(nps_llm_df['processing_ms_val'], errors='coerce').mean()
pcs_xgb_runtime = pd.to_numeric(pcs_xgb_df['processing_ms_val'], errors='coerce').mean()

# Convert ms to seconds if available
nps_llm_runtime_s = nps_llm_runtime / 1000 if pd.notna(nps_llm_runtime) else None
pcs_xgb_runtime_s = pcs_xgb_runtime / 1000 if pd.notna(pcs_xgb_runtime) else None

print(f'Unified avg processing time: {unified_avg_time:.2f}s per transcript')
print(f'NPS_LLM avg processing time: {nps_llm_runtime_s}s per call' if nps_llm_runtime_s else 'NPS_LLM: processing_ms not available')
print(f'PCS_XGB avg processing time: {pcs_xgb_runtime_s}s per call' if pcs_xgb_runtime_s else 'PCS_XGB: processing_ms not available')
```

```python
# --- Build Comparison Table ---

comparison_data = {
    'Pipeline': ['Unified (PCS)', 'Unified (NPS)', 'NPS LLM Driver', 'PCS Predictive (XGBoost)'],
    'Metric Predicted': ['PCS', 'NPS', 'NPS', 'PCS'],
    'Avg Runtime/Transcript (s)': [
        f'{unified_avg_time:.2f}*',
        f'{unified_avg_time:.2f}*',
        f'{nps_llm_runtime_s:.2f}' if nps_llm_runtime_s else 'N/A†',
        f'{pcs_xgb_runtime_s:.2f}' if pcs_xgb_runtime_s else 'N/A†',
    ],
    'Avg Cost/Transcript ($)': [
        'N/A‡',
        'N/A‡',
        f'${nps_llm_cost_per_transcript:.5f}',
        f'${pcs_xgb_cost_per_transcript:.5f}',
    ],
    'Macro-Avg Precision': [
        f'{pcs_precision_unified:.3f}',
        f'{nps_precision_unified:.3f}',
        f'{nps_precision_nps_llm:.3f}',
        f'{pcs_precision_xgb:.3f}',
    ],
}

comparison_table = pd.DataFrame(comparison_data)

print('\n' + '='*80)
print('PIPELINE COMPARISON TABLE')
print('='*80)
display(comparison_table)
print('\n* Unified processing time is for full pipeline (both PCS + NPS combined).')
print('† processing_ms not returned by Azure API; timing is per-batch not per-call.')
print('‡ Unified pipeline per-metric token costs unavailable (no per-call debug_info).\n'
      '  Unified also uses BGE-M3 local embedding compute (no API cost).')
```

## 5. Token Metrics

LLM API token usage statistics per transcript. Unified pipeline reports transcript token counts
(no per-call API token breakdown available). NPS_LLM and PCS_predictive report actual API usage.

**Note:** Unified pipeline also uses BGE-M3 embedding compute (local, not reflected in API tokens).

```python
# --- Token Metrics Table ---

def token_stats(series, name):
    """Compute avg/min/max for a token series."""
    s = pd.to_numeric(series, errors='coerce').dropna()
    return {
        'Metric': name,
        'Mean': f'{s.mean():.0f}',
        'Min': f'{s.min():.0f}',
        'Max': f'{s.max():.0f}',
        'Std Dev': f'{s.std():.0f}',
        'Total (200 calls)': f'{s.sum():,.0f}',
    }

token_rows = []

# Unified: transcript token counts (input to pipeline)
token_rows.append(token_stats(eval_df['unified_token_count'], 'Unified — Transcript Tokens (input)'))

# NPS_LLM: API tokens
token_rows.append(token_stats(nps_llm_df['prompt_tokens'], 'NPS LLM — Prompt Tokens'))
token_rows.append(token_stats(nps_llm_df['completion_tokens'], 'NPS LLM — Completion Tokens'))
token_rows.append(token_stats(nps_llm_df['total_tokens'], 'NPS LLM — Total Tokens'))

# PCS_XGBoost: API tokens (feature extraction call)
token_rows.append(token_stats(pcs_xgb_df['prompt_tokens'], 'PCS Predictive — Prompt Tokens'))
token_rows.append(token_stats(pcs_xgb_df['completion_tokens'], 'PCS Predictive — Completion Tokens'))
token_rows.append(token_stats(pcs_xgb_df['total_tokens'], 'PCS Predictive — Total Tokens'))

token_table = pd.DataFrame(token_rows)

print('\n' + '='*80)
print('TOKEN METRICS TABLE')
print('='*80)
display(token_table)
print('\nNote: Unified pipeline does not expose per-call API token usage.')
print('      "Transcript Tokens" = estimated token count of input text (words × 1.3).')
```

```python
# --- Cache Rate Summary ---
nps_llm_cache_rate = (pd.to_numeric(nps_llm_df['cached_tokens'], errors='coerce') /
                      pd.to_numeric(nps_llm_df['prompt_tokens'], errors='coerce')).dropna()
pcs_xgb_cache_rate = (pd.to_numeric(pcs_xgb_df['cached_tokens'], errors='coerce') /
                      pd.to_numeric(pcs_xgb_df['prompt_tokens'], errors='coerce')).dropna()

cache_data = {
    'Pipeline': ['NPS LLM Driver', 'PCS Predictive (XGBoost)'],
    'Avg Cache Rate': [f'{nps_llm_cache_rate.mean():.1%}', f'{pcs_xgb_cache_rate.mean():.1%}'],
    'Min Cache Rate': [f'{nps_llm_cache_rate.min():.1%}', f'{pcs_xgb_cache_rate.min():.1%}'],
    'Max Cache Rate': [f'{nps_llm_cache_rate.max():.1%}', f'{pcs_xgb_cache_rate.max():.1%}'],
}

cache_table = pd.DataFrame(cache_data)
print('\nPrompt Cache Utilization:')
display(cache_table)
```
